```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# Import the dataset (assuming you have 'market_basket.csv' in your working directory)
retail = pd.read_csv('/Users/adamdanielgreen/Desktop/Business Statistics/Association Rules-20231215/Classeur5b.csv',sep=';')
retail.head()
```

```python
retail = list(retail["V1"].apply(lambda x:x.split(",")))
```

```python
one_hot_transformer = TransactionEncoder()
retail_transform = one_hot_transformer.fit_transform(retail)
```

```python
retail_transform
```

```python
retail = pd.DataFrame(retail_transform,columns=one_hot_transformer.columns_)
retail.head()
```

```python
# Frequent itemset Generation
retail_freq = apriori(retail, min_support = 0.01, use_colnames = True)
retail_freq.sort_values(['support'],ascending=False, inplace = True)
retail_freq
```

```python
# Frequent itemset Generation
frequent_itemsets1 = apriori(retail, min_support=0.01, use_colnames=True)
frequent_itemsets1 = frequent_itemsets1[frequent_itemsets1.itemsets.apply(lambda x: len(x) == 1)]
```

```python
frequent_itemsets1
```

```python
# Top 10 frequent 1-itemsets
top_10_frequent_itemsets1 = frequent_itemsets1.sort_values(by='support', ascending=False).head(10)
top_10_frequent_itemsets1
```

```python
# Frequent 2-itemset Generation
frequent_itemsets2 = apriori(retail, min_support=0.01, use_colnames=True)
frequent_itemsets2 = frequent_itemsets2[frequent_itemsets2.itemsets.apply(lambda x: len(x) == 2)]

# Top 10 frequent 2-itemsets
top_10_frequent_itemsets2 = frequent_itemsets2.sort_values(by='support', ascending=False).head(10)
```

```python
top_10_frequent_itemsets2
```

```python
# Frequent 3-itemset Generation
frequent_itemsets3 = apriori(retail, min_support=0.01, use_colnames=True)
frequent_itemsets3 = frequent_itemsets3[frequent_itemsets3.itemsets.apply(lambda x: len(x) == 3)]

# Top 10 frequent 3-itemsets
top_10_frequent_itemsets3 = frequent_itemsets3.sort_values(by='support', ascending=False).head(10)
```

```python
top_10_frequent_itemsets3
```

```python
frequent_itemsets = apriori(retail, min_support=0.01, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)
```

```python
print(rules)
```

```python
# The top ten rules sorted by the lift
top_10_rules = rules.sort_values(by='lift', ascending=False).head(10)
print(top_10_rules)
```

```python
# Finding subsets of rules containing "COFFEE" items
coffee_rules = rules[rules['antecedents'].apply(lambda x: 'COFFEE' in x)]
print(coffee_rules)
```

```python
# Finding subsets of rules that precede "SUGAR JARS" purchases
sugar_rules = rules[rules['consequents'].apply(lambda x: 'SUGAR' in x)]
print(sugar_rules)
```
